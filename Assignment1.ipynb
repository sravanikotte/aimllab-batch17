```python
#performance  of mathematical operations on the given datasets Multiplication,division,Powerset,substraction
import numpy as np
dataset1=np.array([50,46,72,88])
dataset2=np.array([1,0.1,0.001,0.0001])
product=np.multiply(dataset1,dataset2)
product1=dataset1*dataset2
print(product)
print(product1)
Division=np.divide(dataset1,dataset2)
Moddivision=np.mod(dataset1,dataset2)
Floordivision=np.floor_divide(dataset1,dataset2)
Subtraction=np.subtract(dataset1,dataset2)
print(Division)
print(Floordivision)
print(Subtraction)
```

```python
#implement trigonometric and log functions of a given dataset
import numpy as np
patientage=np.array([65,70,80,85,77])
sin1=np.sin(patientage)
cosine=np.cos(patientage)
Tangent=np.tan(patientage)
log1=np.log10(patientage)
log=np.log2(patientage)
exp=np.exp(patientage)
print(sin1)
print(cosine)
print(Tangent)
print(log1)
print(log)
print(exp)
```

Part-2:Lab Task is to visualize the datasets or data using Matplotlib function graphs

```python
people=['vyshnavi','chen','pranathi','sravani','spandana']
age=[10,25,30,35,45]
weight=[50,40,45,60,80]
height=[150,180,175,160,130]
import matplotlib.pyplot as plt
plt.scatter(weight,height)
plt.title("relationship between weight and height of patients")
plt.ylabel("height in meters")
plt.xlabel("weight of patients in kgs")
plt.show()
```

Generate Bar Plot

```python
plt.figure(figsize=(12,4))
plt.title("people's weight in kgs",fontsize=16,fontstyle='italic')
plt.bar(x=people,height=weight, width=0.6,
color='pink',edgecolor='k',alpha=0.6)
plt.xlabel("People",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("Weight (in kgs)",fontsize=15)
plt.show()
```

Generate Histogram

```python
import numpy as np
plt.figure(figsize=(7,5))
# Main plot function 'hist'
plt.hist(weight,color='black',edgecolor='k', alpha=0.75,bins=5)
plt.title("Histogram of patient weight",fontsize=18)
plt.xlabel("Weight in kgs",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
